> # 상하좌우 반전

### 1. img = cv2.flip(img, 0) : 상하 반전
### 2. img = cv2.flip(img, 1) : 좌우 반전
### 3. img = cv2.flip(img, -1) : 상하좌우 반전

```python
import cv2
from pop import Util

Util.enable_imshow()

haar_face='/usr/local/share/opencv4/haarcascades/haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(haar_face)

cam = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)
if not camera.isOpened():
    print("Not found camera")

while True:
    ret, img = camera.read()
    # 이미지를 상하좌우 반전
    img = cv2.flip(img, -1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces= face_cascade.detectMultiScale(gray, scaleFactor=1.3 ,minNeighbors=1,minSize=(100,100))

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    cv2.imshow('img',img)
    if cv2.waitKey(1) == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```

> # 하체 인식

### Auto car의 차체가 낮기 때문에 하체를 인식하는 것으로 변경  
### xml 하체 데이터 : haarcascade_lowerbody.xml

```python
import cv2
from pop import Util

Util.enable_imshow()

haar_face='/usr/local/share/opencv4/haarcascades/haarcascade_lowerbody.xml'
face_cascade = cv2.CascadeClassifier(haar_face)

cam = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

if not camera.isOpened():
    print("Not found camera")

while True:
    ret, img = camera.read()
    # 이미지를 상하 반전
    img = cv2.flip(img, 0)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces= face_cascade.detectMultiScale(gray, scaleFactor=1.1 ,minNeighbors=1,minSize=(100,100))

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        
    cv2.imshow('img',img)
    
    if cv2.waitKey(1) == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```

## 촬영 협조 : 조소양 👍👍👍

### 1. 하체가 인식된 영역의 왼쪽 위 꼭지점 좌표 (x,y)와 해당 영역의 너비(w)와 높이(h)를 구한다.
### 2. 사각형 영역의 중심점과 이미지의 중심점을 비교하여 진행 각도 결정 후 steering 값에 적용한다.
### 3. 적용된 steering 값을 이용하여 차량 이동 제어를 한다.

```python
import cv2
from pop import Pilot, Util
import time

Util.enable_imshow()

haar_face='/usr/local/share/opencv4/haarcascades/haarcascade_lowerbody.xml'
face_cascade = cv2.CascadeClassifier(haar_face)

car = Pilot.AutoCar()

cam = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

if not camera.isOpened():
    print("Not found camera")

while True:
    ret, img = camera.read()
    # 이미지를 상하 반전
    img = cv2.flip(img, -1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces= face_cascade.detectMultiScale(gray, scaleFactor=1.1 ,minNeighbors=1,minSize=(100,100))

    if len(faces) > 0:
        (x,y,w,h) = faces[0]
        # 얼굴 중앙점의 위치 계산
        center_x = x + w // 2
        # 이미지 너비의 중앙점
        img_center_x = img.shape[1] // 2
        # 중앙점 간 거리 계산
        distance = img_center_x - center_x
        # 거리에 따라 조향각 설정
        angle = -distance / (img_center_x / 2)

        car.steering = angle
        car.setSpeed(60)
        car.forward()
        
    else:
        # 얼굴을 인식하지 못한 경우 정지
        car.stop()

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

    cv2.imshow('img',img)

    if cv2.waitKey(1) == ord('q'):
        break

car.stop()
camera.release()
cv2.destroyAllWindows()
```

# 문제 발생
### 1. angle 값이 car.steering이 가질 수 있는 -1 ~ 1 값을 벗어나는 경우가 존재
### 2. 차량 이동시 steering 값이 적용되고 다음 steering 값이 적용되기까지 시간이 김 - 차량이 S 자로 주행

> # 최종코드 : AutoCar 이동 제어 추가

# 해결
### 1. angle 값이 -1보다 작으면 -1, 1보다 크면 1값을 적용
### 2. time.sleep() 을 이용하여 짧은 시간동안만 steering 값이 적용되게 변경

```python
import cv2
from pop import Pilot, Util
import time

Util.enable_imshow()

haar_face='/usr/local/share/opencv4/haarcascades/haarcascade_lowerbody.xml'
face_cascade = cv2.CascadeClassifier(haar_face)

car = Pilot.AutoCar()

cam = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

if not camera.isOpened():
    print("Not found camera")

while True:
    ret, img = camera.read()
    # 이미지를 상하 반전
    img = cv2.flip(img, -1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces= face_cascade.detectMultiScale(gray, scaleFactor=1.1 ,minNeighbors=1,minSize=(100,100))

    if len(faces) > 0:
        (x,y,w,h) = faces[0]
        # 얼굴 중앙점의 위치 계산
        center_x = x + w // 2
        # 이미지 너비의 중앙점
        img_center_x = img.shape[1] // 2
        # 중앙점 간 거리 계산
        distance = img_center_x - center_x
        # 거리에 따라 조향각 설정
        angle = -distance / (img_center_x / 2)
        # 조향각 적용
#         print(angle)
        if angle < -1:
            angle = -1
        elif angle > 1:
            angle = 1
        else:
            angle = angle
            
        car.steering = angle
        car.setSpeed(60)
        car.forward()
        
        time.sleep(0.3)
        car.steering = 0
        car.setSpeed(60)
        car.forward()
        
    else:
        # 얼굴을 인식하지 못한 경우 정지
        car.stop()

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

    cv2.imshow('img',img)

    if cv2.waitKey(1) == ord('q'):
        break

car.stop()
camera.release()
cv2.destroyAllWindows()
```
