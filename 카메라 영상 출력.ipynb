> <h2> 카메라 캡처 코드 활용

- 사각형, 원, 텍스트 코드 추가

```python
import cv2
from pop import Util

Util.enable_imshow()

cam = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

if not camera.isOpened():
    print("Not found camera")
    
width = camera.get(cv2.CAP_PROP_FRAME_WIDTH)
height = camera.get(cv2.CAP_PROP_FRAME_HEIGHT)

print("init width: %d, init height: %d" % (width,height))

for _ in range(120):
    ret, frame = camera.read()
    if not ret:
        break


    cv2.rectangle(frame, (50, 50), (250, 250), color=(255, 255, 0), thickness=2)     # 사각형 : 도형 위치, 크기, 색깔, 두께 설정
    cv2.circle(frame, (150, 350), 100, color=(255, 0, 255), thickness=2)             # 원 : 도형 위치, 크기, 색깔, 두께 설정
    font =  cv2.FONT_HERSHEY_PLAIN                                                   # 폰트 설정
    #blue = (255, 0, 0)                                                              # 색깔 설정
    cv2.putText(frame, "Pretzels", (350, 100), font, 3, (255, 0, 0), 3, cv2.LINE_AA) # 텍스트 입력 : 내용, 위치, 폰트, 폰트 크기, 색깔, 두께 설정, 선의 종류
    
    cv2.imshow("soda", frame)

camera.release()
cv2.destroyAllWindows()
```

> <h2> 결과

> <h2> opencv_jupyter_ui 라이브러리 활용

```python
import opencv_jupyter_ui as jcv2
from pop import Util

Util.enable_imshow()
cam = Util.gstrmer(width=640, height=480)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

if (camera.isOpened() == False):
    print("Unable to open camera")
    
while True:
    ret, frame = camera.read()
    if (not ret) or (jcv2.waitKey(1) == ord('q')): 
        break
        
    cv2.rectangle(frame, (50, 50), (250, 250), color=(255, 255, 0), thickness=2)      # 사각형 : 도형 위치, 크기, 색깔, 두께 설정
    cv2.circle(frame, (150, 350), 100, color=(255, 0, 255), thickness=2)              # 원 : 도형 위치, 크기, 색깔, 두께 설정
    font =  cv2.FONT_HERSHEY_PLAIN                                                    # 폰트 설정
    cv2.putText(frame, "Pretzels", (350, 100), font, 3, (255, 0, 0), 3, cv2.LINE_AA)  # 텍스트 입력 : 내용, 위치, 폰트, 폰트 크기, 색깔, 두께 설정, 선의 종류
    
    cv2.imshow('Video', frame)
    jcv2.imshow('Video', frame)
    
camera.release()
jcv2.destroyAllWindows()
```
